```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

# 设置Chrome浏览器的驱动程序路径
chromedriver_path = '"C:\Program Files\Google\Chrome\Application\chrome.exe"'

# 设置网站的URL和待采集的图表元素的CSS选择器
url = "https://bitinfocharts.com/comparison/bitcoin-transactionfees.html#alltime"
#url = 'https://bitinfocharts.com/comparison/google_trends-btc.html#alltime'
#url = 'https://bitinfocharts.com/comparison/tweets-btc.html#alltime'
#url = 'https://bitinfocharts.com/comparison/transactionvalue-btc.html#alltime'
#url = 'https://bitinfocharts.com/comparison/bitcoin-marketcap.html#alltime'
#url = 'https://bitinfocharts.com/comparison/sentinusd-btc.html#alltime'
#url = 'https://bitinfocharts.com/comparison/bitcoin-transactions.html#alltime'
#url = 'https://bitinfocharts.com/comparison/activeaddresses-btc.html#alltime'
chart_selector = '#chart'

# 启动Chrome浏览器



# 创建一个空的DataFrame来存储数据
data = pd.DataFrame(columns=['x', 'y'])

# 循环遍历图表的横坐标
X = 1690
Y = 300
x_value = '2023/3/3'
for i in range(1,15):
    browser = webdriver.Chrome()
    browser.maximize_window() 
    browser.get(url)
    browser.implicitly_wait(5)
    
    ActionChains(browser).move_by_offset(X-120*(i-1),Y).perform()
    ActionChains(browser).click_and_hold().perform()
    time.sleep(0.5)
    ActionChains(browser).move_by_offset(-120,0).perform()
    time.sleep(0.5)
    ActionChains(browser).release().perform()
    time.sleep(1.5)
    ActionChains(browser).move_by_offset(120*i,0).perform()
    time.sleep(0.5)
    page = browser.page_source
    for j in range(1,610):
        # 模拟鼠标在图表上移动到横坐标x
        try: ActionChains(browser).move_by_offset(-3,0).perform()
        except: break
        browser.implicitly_wait(0.05)
        # 等待一会儿，直到提示框出现
        text = browser.find_elements(By.CLASS_NAME,"dygraph-legend")[-1].text
        try:
            date = re.search(r'^[0-9]{4}/[0-9]{2}/[0-9]{2}',text).group()
            y_value = re.search(r'[0-9]*\.?[0-9]*[A-Za-z]?$',text).group()
            print(date,y_value)
        except: pass
        if (date < x_value):
                x_value = date
                data.loc[len(data.index)]=[x_value,y_value]
        else:
            continue
        # 获取提示框的HTML内容
        #date = re.compile(r'(?<=div class="dygraph-legend").*(?=>").+?(?=:"<br>)',re.S)
        #value = re.compile(r'(?<=":&nbsp;).+?(?="[\s\S])',re.S)
        #temp = re.findall(date,page)
        #y_value = re.findall(value,page)
        #print(temp)
        #print(y_value)
    browser.close()
    browser.quit()          



# 将DataFrame保存为CSV文件
data.to_csv('chart_data.csv', index=False)
```
